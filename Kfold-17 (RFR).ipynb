```python
from sklearn.svm import SVR
from sklearn.model_selection import RepeatedKFold,ShuffleSplit
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import RFE 
from sklearn import metrics
d={0:0,
  5:0,
  10:0,
    100:0
}
cv = RepeatedKFold(n_splits=5, n_repeats=20, random_state=123)
count1RF=0
count2RF=0
maxR2=0
selector = RandomForestRegressor(max_depth= 21,n_estimators=36, random_state=False)
for train, test in cv.split(y):
    selector = selector.fit(X, y)
    y_pred2 = selector.predict(X[test])
    x_pred2 = selector.predict(X[train])
    R2test2=1-(((y[test] - y_pred2) ** 2).sum())/((y[test] - y[test].mean()) ** 2).sum()
    R2train2=1-(((y[train] - x_pred2) ** 2).sum())/((y[train] - y[train].mean()) ** 2).sum()
    if abs(R2train2-R2test2)*100>10:
        d[100]+=1
    elif abs(R2train2-R2test2)*100>5:
        d[10]+=1
    elif abs(R2train2-R2test2)*100>1:
        d[5]+=1
    else:
        d[0]+=1
    count1RF+=R2test2
    count2RF+=R2train2
    maxR2=max(maxR2,R2test2)
    if R2test2>0.95:
        plt.show()
        print("------------------------------------------------------------")
        row = 1
        col = 0
        worksheet.write(row, 0, 'test number')
        worksheet.write(row, 1, 'real value')
        worksheet.write(row, 2, 'predicted value')
        for num, real, pred in zip(test,y[test],y_pred2):
            worksheet.write(row, col,     num)
            worksheet.write(row, col + 1, real)
            worksheet.write(row, col + 2, pred)
            row += 1
        worksheet.write(row, 0, 'Train R2')
        worksheet.write(row, 1, R2train2)
        worksheet.write(row+1, 0, 'Test R2')
        worksheet.write(row+1, 1, R2test2)
        mae = metrics.mean_absolute_error(y[test], y_pred2)
        worksheet.write(row+2, 0, 'Mean absolute error')
        worksheet.write(row+2, 1, mae)
        
        print(test)
        print("predicted Value in test set with RandomForest:")
        print(y_pred2)
        print("Test accuracy result in RandomForest:",R2test2)
        print('Mean Absolute Error in RandomForest:', mae)
        print("Train accuracy result in RandomForest:",R2train2)
        print("------------------------------------------------------------")
print(maxR2)
print(X[0,:])
print(d)
print(selector.feature_importances_)
workbook.close()
#print(selector.support_)
#print(selector.ranking_)
print("Test accuracy average in RandomForest:",count1RF/100.0)
print("Train accuracy average in RandomForest:",count2RF/100.0)
```
