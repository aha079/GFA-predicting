```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("Final Dataset.csv",usecols = ['Tx','Tg','Tl','Y'])
dataset.head(5)
```

```python
X = dataset.iloc[:,0:3].values.astype(float)
X[:,0] = (X[:,0] - min(X[:,0]))/(max(X[:,0])-min(X[:,0]))
X[:,1] = (X[:,1] - min(X[:,1]))/(max(X[:,1])-min(X[:,1]))
X[:,2] = (X[:,2] - min(X[:,2]))/(max(X[:,2])-min(X[:,2]))
y = dataset.iloc[:,3].values.astype(float)
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import RepeatedKFold,ShuffleSplit
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
count1=0
count1RF=0
count2=0
count2RF=0
rkf = ShuffleSplit(n_splits=10000, test_size=0.2, random_state=12432)
regressor = SVR(kernel='poly', C=70, gamma='auto', degree=3, epsilon=0.5,coef0=1)
regressor2 = RandomForestRegressor(max_depth= 12,n_estimators=45, random_state=False)
for train, test in rkf.split(y):
    regressor.fit(X[train],y[train])
    y_pred = regressor.predict(X[test])
    x_pred = regressor.predict(X[train])
    regressor2.fit(X[train],y[train])
    y_pred2 = regressor2.predict(X[test])
    x_pred2 = regressor2.predict(X[train])
    R2test=1-(((y[test] - y_pred) ** 2).sum())/((y[test] - y[test].mean()) ** 2).sum()
    R2test2=1-(((y[test] - y_pred2) ** 2).sum())/((y[test] - y[test].mean()) ** 2).sum()
    R2train=1-(((y[train] - x_pred) ** 2).sum())/((y[train] - y[train].mean()) ** 2).sum()
    R2train2=1-(((y[train] - x_pred2) ** 2).sum())/((y[train] - y[train].mean()) ** 2).sum()
    count1+=R2test
    count1RF+=R2test2
    count2+=R2train
    count2RF+=R2train2
    if R2test2>0.65 or R2test>0.6:
        importance = regressor2.feature_importances_
        plt.bar(['Tx','Tg','Tl'], importance)
        plt.xlabel("Feature Importance")
        plt.show()
        print("------------------------------------------------------------")
        print(test)
        print("predicted Value in test set with SVR:")
        print(y_pred)
        print("predicted Value in test set with RandomForest:")
        print(y_pred2)
        print("Test accuracy result in RandomForest:",R2test2)
        print("Test accuracy result in SVR:",R2test)
        print('Mean Absolute Error in RandomForest:', metrics.mean_absolute_error(y[test], y_pred2))
        print('Mean Absolute Error in SVR:', metrics.mean_absolute_error(y[test], y_pred))
        print("Train accuracy result in RandomForest:",R2train2)
        print("Train accuracy result in SVR:",R2train)
        print("------------------------------------------------------------")
print("Test accuracy average in SVR:",count1/10000.0)
print("Test accuracy average in RandomForest:",count1RF/10000.0)
print("Train accuracy average in SVR:",count2/10000.0)
print("Train accuracy average in RandomForest:",count2RF/10000.0)
```
