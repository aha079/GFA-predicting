```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.inspection import permutation_importance
```

```python
dataset = pd.read_csv("Book1.csv",usecols = ['Tx','Tl','Tg','X1','X2','X3','X5','X6','X7','X8','X10','X11','X12','X13','Y'])
dataset.head(5)
```

```python
X = dataset.iloc[:,0:14].values.astype(float)
X[:,0] = (X[:,0] - min(X[:,0]))/(max(X[:,0])-min(X[:,0]))
X[:,1] = (X[:,1] - min(X[:,1]))/(max(X[:,1])-min(X[:,1]))
X[:,2] = (X[:,2] - min(X[:,2]))/(max(X[:,2])-min(X[:,2]))
X[:,3] = (X[:,3] - min(X[:,3]))/(max(X[:,3])-min(X[:,3]))
X[:,4] = (X[:,4] - min(X[:,4]))/(max(X[:,4])-min(X[:,4]))
X[:,5] = (X[:,5] - min(X[:,5]))/(max(X[:,5])-min(X[:,5]))
X[:,6] = (X[:,6] - min(X[:,6]))/(max(X[:,6])-min(X[:,6]))
X[:,7] = (X[:,7] - min(X[:,7]))/(max(X[:,7])-min(X[:,7]))
X[:,8] = (X[:,8] - min(X[:,8]))/(max(X[:,8])-min(X[:,8]))
X[:,9] = (X[:,9] - min(X[:,9]))/(max(X[:,9])-min(X[:,9]))
X[:,10] = (X[:,10] - min(X[:,10]))/(max(X[:,10])-min(X[:,10]))
X[:,11] = (X[:,11] - min(X[:,11]))/(max(X[:,11])-min(X[:,11]))
X[:,12] = (X[:,12] - min(X[:,12]))/(max(X[:,12])-min(X[:,12]))
X[:,13] = (X[:,13] - min(X[:,13]))/(max(X[:,13])-min(X[:,13]))

y = dataset.iloc[:,14].values.astype(float)
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import RepeatedKFold,ShuffleSplit
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
rkf = ShuffleSplit(n_splits=40000, test_size=0.2, random_state=12432)
count1RF=0
count2RF=0
regressor2 = RandomForestRegressor(max_depth= 4,n_estimators=15, random_state=False)
for train, test in rkf.split(y):
    regressor2.fit(X[train],y[train])
    y_pred2 = regressor2.predict(X[test])
    x_pred2 = regressor2.predict(X[train])
    R2test2=1-(((y[test] - y_pred2) ** 2).sum())/((y[test] - y[test].mean()) ** 2).sum()
    R2train2=1-(((y[train] - x_pred2) ** 2).sum())/((y[train] - y[train].mean()) ** 2).sum()
    count1RF+=R2test2
    count2RF+=R2train2
    if R2test2>0.9:
        perm_importance = permutation_importance(regressor2, X[test], y[test])
        plt.barh(['Tx','Tl','Tg','X1','X2','X3','X5','X6','X7','X8','X10','X11','X12','X13'], perm_importance.importances_mean)
        plt.xlabel("Permutation Importance")
        plt.show()
        importance = regressor2.feature_importances_
        plt.bar(['Tx','Tl','Tg','X1','X2','X3','X5','X6','X7','X8','X10','X11','X12','X13'], importance)
        plt.xlabel("Feature Importance")
        plt.show()
        print("------------------------------------------------------------")
        print(test)
        print("predicted Value in test set with RandomForest:")
        print(y_pred2)
        print("Test accuracy result in RandomForest:",R2test2)
        print('Mean Absolute Error in RandomForest:', metrics.mean_absolute_error(y[test], y_pred2))
        print("Train accuracy result in RandomForest:",R2train2)
        print("------------------------------------------------------------")
print("Test accuracy average in RandomForest:",count1RF/40000.0)
print("Train accuracy average in RandomForest:",count2RF/40000.0)
```
