```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
a=1
t=np.linspace(0.0,2*np.pi,200)
x=np.cos(t)
y=np.sin(t)
plt.scatter(y,x,color='g')
plt.annotate("0_0", xy=(1, 1), xytext=(-0.05, 0),arrowprops=dict(arrowstyle="->"))
plt.show()
points=(x,y)
```

```python
def pi_controller(x,y,points,head,vel,kp,ki,kteta,iteration=30000,d_star=0.1,dt=0.02):
    x_path=[]
    y_path=[]
    heads=[]
    vels=[]
    errors =[] 
    point_x,point_y=points
    integ=0
    #broad casting
    delta_x = point_x - x
    delta_y = point_y - y
    dis=np.sqrt(np.square(delta_x)+np.square(delta_y))
    bestpoint=np.argmin(dis)
    best_x = round(point_x[bestpoint],6)
    best_y = round(point_x[bestpoint],4)
    for i in range(iteration):
        x_path.append(x)
        y_path.append(y)
        heads.append(head)
        vels.append(vel)
        
        delta_x = best_x-x
        delta_y = best_y-y
        
        dis_err = np.sqrt(np.square(delta_x)+np.square(delta_y))
        vel = kp*dis_err+ ki*integ
        integ+=dis_err*dt
        errors.append(dis_err)
        best_teta = math.atan2(delta_y,delta_x)
        head = (vel*20) * np.tan(kteta*(best_teta-head))
        x+= vel*np.cos(head) * dt
        y+= vel*np.sin(head) * dt
        if dis_err <= d_star:
            d_star = dis_err
            bestpoint = (bestpoint+1)%len(point_x)
            best_x = round(point_x[bestpoint],6)
            best_y = round(point_y[bestpoint],4)
    return (x_path,y_path,heads,vels,errors)
generator = pi_controller(0,0,points,math.atan2(1,1),1,kp=0.1,ki=0.001,kteta=0.05)
```

```python
plt.figure(figsize=(9, 16))
plt.plot(y, x,color='r')
plt.scatter(generator[0][:] , generator[1][:] , color='b' )
plt.annotate("start", xy=(np.sin(generator[2][0]), np.cos(generator[2][0])), xytext=(generator[0][0] , generator[1][0]),fontsize=20)
plt.show()
```

```python

plt.plot([x * 0.02 for x in range(len(generator[4]))],generator[4], label='Error')
plt.xlabel('Time ($second$)', fontsize=16)
plt.ylabel('Error', fontsize=16)
plt.show()
```
